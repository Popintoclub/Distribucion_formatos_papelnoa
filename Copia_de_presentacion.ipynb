## Resolución matricial
la resolución matricial se realiza atravez de la lista de vectores "posibles" que son las posibles combinaciones de formatos aplicables 
en bobinadora en donde su suma se encuentra entre los valores [340,383] y la lista de formatos, se crea una matriz extendida de la 
siguiente forma agregándole una ultima columna con los valores de "banda" para esas combinaciones se lo calcula haciendo 383-sumatoria 
de formatos con las siguientes condiciones.
1. si la suma de los formatos es mayor o igual a 370 cm no se agrega banda y se reduce el ancho.
2. Si la suma de los formatos es menor a 370 cm se agrega banda hasta completar 383 cm.

\begin{equation}
\begin{matrix}
vector & formato(95) & formato(145) & formato(150) & formato(200) & formato(213) & posibles & banda\\
v1 & 0 & 1 & 0 & 0 & 1 & (213, 145) & 25\\
v2 & 0 & 0 & 1 & 0 & 1 & (213, 150) & 20\\
v3 & 0 & 1 & 0 & 1 & 0 & (200, 145) & 38\\
v4 & 0 & 0 & 1 & 1 & 0 & (200, 150) & 33\\
v5 & 2 & 0 & 1 & 0 & 0 & (150, 95, 95) & 43\\
v6 & 4 & 0 & 0 & 0 & 0 & (95, 95, 95, 95) & 0\\
\end{matrix}
\end{equation}

Resuelve la matriz identidad guardando del lado izquierdo las transformaciones vectoriales.
en la ultima linea agrego las cantidades de bobinas que necesito de cada formato, esto calculado en la primer función del programa.

\begin{equation}
\begin{matrix}
Transformacion & formato(95) & formato(145) & formato(150) & formato(200) & formato(213)\\
\frac{v_{6}}{4} & 1 & 0 & 0 & 0 & 0\\
v_{1}-(v_{2}-v_{5}-\frac{v_{6}}{2}) & 0 & 1 & 0 & 0 & 0\\
v_{5}-\frac{v_{6}}{2} & 0 & 0 & 1 & 0 & 0\\
v_{4}-(v_{5}-\frac{v_{6}}{2}) & 0 & 0 & 0 & 1 & 0\\
v_{2}-(v_{5}-\frac{v_{6}}{2}) & 0 & 0 & 0 & 0 & 1\\
v_{3}-(v_{1}-v_{2}-v_{5}-\frac{v_{6}}{2}) & 0 & 0 & 0 & 1 & 0\\
cantidad: & 8 & 8 & 4 & 4 & 8\\
\end{matrix}
\end{equation}$

## Encontrando las posibles ordenes de produccion:  
existe un vector que es linealmente dependiente a otro entonces se generan dos bases al intercambiar esos vectores. es decir que para una 
matriz de $m*n$ van a existir $m-n$ bases canonicas posibles, en el ejemplo quee stamos llevando son $m$ valores posibles y $n$ formatos
por lo tanto van a existir dos posibles ordenes de produccion.

Se multiplican las transformaciones por la base que se utiliza y sumar los valores:    

***Orden de prodccion 1:***    

utiliza los 5 vectores que forman la base canonica. 
\begin{equation}
\begin{matrix}
Transformacion & formato(95) & formato(145) & formato(150) & formato(200) & formato(213)\\
\frac{v_{6}}{4} & 1 & 0 & 0 & 0 & 0\\
v_{1}-v_{2}+v_{5}-\frac{v_{6}}{2} & 0 & 1 & 0 & 0 & 0\\
v_{5}-\frac{v_{6}}{2} & 0 & 0 & 1 & 0 & 0\\
v_{4}-v_{5}+\frac{v_{6}}{2} & 0 & 0 & 0 & 1 & 0\\
v_{2}-v_{5}+\frac{v_{6}}{2} & 0 & 0 & 0 & 0 & 1\\
cantidad: & 8 & 8 & 4 & 4 & 8\\
\end{matrix}
\end{equation}  

simplemente tenemos que multiplicar la cantidad de bobinas que necesitamos por cada transformacion
  
$ (\frac{v_{6}}{4})*8 + (v_{1}-v_{2}+v_{5}-\frac{v_{6}}{2})*8 +(v_{5}-\frac{v_{6}}{2})*4 + (v_{4}-v_{5}+\frac{v_{6}}{2})*4 +(v_{2}-v_{5}+\frac{v_{6}}{2})*8$
   
por lo tanto realizando la multiplicacion de cantidad de bobinas por las transformaciones y sumandolas no da:  
    

  
$2*v_{6}+8*v_{1}+4*v_{4}$  
  
esto implica la orden de produccion de la forma: 
  
  
$[bajadas] *[formatos]$  
  
  
$2*[95,95,95,95]+8*[213,145]+4*[150,200]$  
   
  
**Orden de produccion 2: **   

utiliza los 5 vectores que forman la base canonica intrcambiando el linealmente independiente. 
\begin{equation}
\begin{matrix}
Transformacion & formato(95) & formato(145) & formato(150) & formato(200) & formato(213)\\
\frac{v_{6}}{4} & 1 & 0 & 0 & 0 & 0\\
v_{1}-v_{2}+v_{5}-\frac{v_{6}}{2} & 0 & 1 & 0 & 0 & 0\\
v_{5}-\frac{v_{6}}{2} & 0 & 0 & 1 & 0 & 0\\
v_{3}-v_{1}+v_{2}-v_{5}+\frac{v_{6}}{2} & 0 & 0 & 0 & 1 & 0\\
v_{2}-v_{5}+\frac{v_{6}}{2} & 0 & 0 & 0 & 0 & 1\\
cantidad: & 8 & 8 & 4 & 4 & 8\\
\end{matrix}
\end{equation}  
  
$(\frac{v_{6}}{4})*8+(v_{1}-v_{2}+v_{5}-\frac{v_{6}}{2})*8+(v_{5}-\frac{v_{6}}{2})*4+(v_{3}-v_{1}+v_{2}-v_{5}+\frac{v_{6}}{2})*4+(v_{2}-v_{5}+\frac{v_{6}}{2})*8$  

Resumiendo la O.P. es

$2*v_{6}+4*v_{1}+4*v_{2}+4*v_{3}$  
  
$[bajadas] *[formatos]$  
  
$2*[95,95,95,95] + 4*[213,145] + 4*[213,150] +*[200,145]$
  
    
De esta forma, el programa encuentra todas las posibles ordenes de produccion, suma las pastas de cada combinacion multiplicandola 
por la cantidad de bajadas y selecciona la orden de produccion con la menor banda total producida.   
  
## Optimizando las ordenes de produccion a travez de un Error inducido.


No siempre los resultados van a ser numeros enteros, ya que al haber varias bobinas de un mismo corte para obtener la canonica se lo 
divide por el mismo numero, ejemplo [125,125,125] para utilizar ese vector vamos a tener que dividirlo por 3, es por eso que aplico
un error E que se lo puede agruegar en el entorno grafico y que significa la variacion de posibles ordenes de produccion aumentando 
cantidad de bobinas a producir.  
  
\begin{equation}
\begin{matrix}
Transformacion & formato(95) & formato(145) & formato(150) & formato(200) & formato(213)\\
\frac{v_{6}}{4} & 1 & 0 & 0 & 0 & 0\\
v_{1}-(v_{2}-v_{5}-\frac{v_{6}}{2}) & 0 & 1 & 0 & 0 & 0\\
v_{5}-\frac{v_{6}}{2} & 0 & 0 & 1 & 0 & 0\\
v_{4}-(v_{5}-\frac{v_{6}}{2}) & 0 & 0 & 0 & 1 & 0\\
v_{2}-(v_{5}-\frac{v_{6}}{2}) & 0 & 0 & 0 & 0 & 1\\
v_{3}-(v_{1}-v_{2}-v_{5}-\frac{v_{6}}{2}) & 0 & 0 & 0 & 1 & 0\\
cantidad: & 8+E & 8+E & 4+E & 4+E & 8+E\\
\end{matrix}
\end{equation}  

  
Los datos de salida se exportan a excel las tablas de datos programados y la orden de produccion seleccionada, Es decir el programa 
calcula las ordenes de produccion posibles con el lote de bobinas seleccionadas y una vez obtenidas elige cual entregar en base a lo 
que el programador le pida, en el ejemplo anterior la condicion fue que seleccione la O.P. que haya generado menos "banda"
